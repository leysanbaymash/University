# 2. Изоморфизм линейных пространств.

Два линейных пространства $ V_1, V_2 $ над обзим полем $ P $ называются изоморфными, если существует биективное отображение $ \varphi: V_1 \to V_2 $, которое сохраняет законы композиции: $ \forall x, y \in V_1, \space \alpha \in P $
1. $ \varphi(x + y) = \varphi(a) + \varphi(b) $
2. $ \varphi(\alpha x) = \alpha \varphi(a) $  

Обозначение: $ V_1 \cong V_2 $

Примеры
1. $ V_n \cong \mathbb{R}^n $
2. $ V_2 \cong C_\mathbb{R} $
3. $ P^{n \times m} \cong P^{nm} $

Простейшие свойства изоморфных пространств:
1. Отношение изоморфизма есть отношение эквивалентности на множестве всех линейных пространств над полем $ P $
2. В изоморфных пространствах
    - образ и прообраз линейное комбинации векторов есть линейная комбинация образов и прообразов с теми же коэффицентами
    - образ и прообраз нулевого вектора есть нулевой вектор
    - образ и прообраз линейно независимой системы векторов образует линейно независимую систему векторов
    - образ и прообраз базиса есть базис

Доказательства опираются на определения и свойства рассматриваемых структур.

### Теорема 63.1 (критерий изоморфизма)

Два линейных пространства над общим полем изоморфны тогда и только тогда, когда их размерности совпадают.

### Доказательство

Необходимость вытекает из свойства, связанного с базисами.

Достаточность. Пусть $ \dim V_1 = \dim V_2 = n $. Выберем для этих пространств базисы $ e_1, \dots, e_n $ и $ f_1, \dots, f_n $ соотвественно. Построим обозражение $ \varphi: V_1 \to V_2 $. Каждому вектору $ x = \sum\limits_{i = 1}^n \alpha_i e_i \space \in V_1 $ поставим в соотвествие вектор $ y = \sum\limits_{i = 1}^n \alpha_i f_i \space \in V_2 $. В силу единственности разложения вектора по базису, отображение биективно и эти пространства изоморфны.

Следствие. Любое $ n $ - мерное вещественное пространство изоморфно арифметическому пространству $ \mathbb{R}^n $, а любое $ n $ - мерное комплексное пространство - арифметическому пространству $ \mathbb{C}^n $.

Изоморфизм, построенный при доказательстве теоремы - изоморфизм, переводящий базис в базис - называется координатным изоморфизмом, определённым базисом $ e_1, \dots, e_n $. Координатный изоморфизм "заменяет" линейное пространство конкретным арифметическим пространством и сводит решение многих задач к вычислениям в поле.
