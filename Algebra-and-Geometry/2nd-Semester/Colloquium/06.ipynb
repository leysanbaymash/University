### Теорема 70.1

Ортогональная система ненулевых векторов линейно независива.

### Доказательство

Полагаем, что она зависима, записываем эту систему с коэффиценами такими, чтобы линейная комбинация была равна нулю, и умножаем её на $ x_i $. Получем $ \alpha (x_i, x_i) = 0 \Rightarrow x_i = \theta $, что противоречит условию.

Следствия:
1. Любая ортонормированная система векторов линейно независима
2. В n-мерном пространстве любая ортонормированная система из $ n $ векторов является базисом.

### Теорема 70.4

В конечномерном евклидовом (унитарном) пространстве существует ортонормированный базис.

### Доказательство

Пусть $ \dim V = n $. Используем индукцию по $ n $. Пусть в любом евклидовом подпространстве размерности $ n - 1 $ существует ортонормированный базис. Покажем, что ортонормированный базис существует и в пространстве размерности $ n $. Рассмотрим базис пространства размерности $ n \space f_1, \dots, f_n $, возьмём подпространство $ L(f_1, \dots, f_{n - 1}) $, для него зададим ортонормированный базис $ e_1, \dots, e_{n - 1} $. Потом зададим вектор $ g_n = f_n - \alpha_1 e_1 - \dots - \alpha_{n - 1} e_{n - 1} $. Положим $ 0 = (g_n, e_i) = (f_n - e_i) - \alpha_i, \space i = \overline{1, n} $. Базис $ e_1, \dot, e_{n - 1}, g_n $ ортогонален. Теперь положим $ e_n = \frac{g_n}{|g_n|} $. Мы получили ортонормированный базис.
